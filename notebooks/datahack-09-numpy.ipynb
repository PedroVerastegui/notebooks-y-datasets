## comparación de tiempos de ejecución

```python
lista = list(range(100000000))
numpy_array = numpy.array(lista)
```

```python
def suma(lista):
    suma = 0
    for n in lista:
        suma = suma + n
    return suma
```

```python
%timeit suma(lista)
```

```python
%timeit sum(lista)
```

```python
%timeit np.sum(numpy_array)
```
