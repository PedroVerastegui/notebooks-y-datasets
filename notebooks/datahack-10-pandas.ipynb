## apply

```python
df_pokedex['generation'].unique()
```

```python
def generaciones(numero):
    if numero == 1:
        return 'Kanto'
    elif numero == 2:
        return 'Johto'
    elif numero == 3:
        return 'Hoenn'
    elif numero == 4:
        return 'Sinnoh'
    elif numero == 5:
        return 'Unova'
    elif numero == 6:
        return 'Kalos'
    elif numero == 7:
        return 'Alola'
```

```python
df_pokedex['region'] = df_pokedex['generation'].apply(generaciones)
```

```python
df_pokedex.head()
```

```python
df_pokedex.tail()
```
